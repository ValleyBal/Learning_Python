```python
#Program 1 
#imporing math to use the functions to calculate with pi
import math
#setting num1 as an int to tell python it will be a number. 
#input is listed to give prompt for number to enter.
num1= int( input("Enter radius ") )
#setting variable to calculate diameter of circle. 
diameter=round(2*num1,2)
#setting variable to calculate area of circle. 
#using math.pi in place of the actual pi value.
#using the round function to limit the digits to 2 decimal places.
area=round(math.pi*num1*num1,2)
#print function with the variables added to form a sentence.
print("The diameter of the circle is",diameter,"and the area of the circle is", area)
```

```python
#Program 2
#setting the variable as a number with int and added input function to enter number. 
celsius = int(input("Enter degrees in Celcius:"))
#variable is set to calculate at the end of the printed string.
print("The degree when converted to Fahrenheit is:", celsius*1.8+32)
```

```python
#Program 3
#made input varaibles to ingest the numbers needed for the calculation.
wrkHours = int(input("What is your total work hours for the week?"))
payrateHr = float(input("What is your pay rate per hour?"))
#subtracted to isolate the overtime pay.
ovrTime = wrkHours-40
#multiplied regular pay with 1.5 to get 50% higher wage rate.
ovrPay = payrateHr*1.5
#multiplied the overtime pay rate with the overtime hours. 
ovrTimePay = ovrTime*ovrPay
#created if functions to calculate pay if less than or equal to 40 hours. 
if wrkHours<=40:
    weeklypay=payrateHr*wrkHours
#created elif function to calculate the variables if overtime pay exceeded 40 hours. 
#calculations were made in lines above.
elif wrkHours>40:
    weeklypay=payrateHr*wrkHours+ovrTimePay
#printed important sections of the calculation to ensure clarity. 
print("Overtime Pay Rate:",ovrPay)    
print("Your overtime hours:",ovrTime)
print("Your weekly overtime pay is:",ovrTimePay)
print("Your weekly gross pay total is:",weeklypay)
```

```python
#Program 4
#created an input function with num1 to ingest the numbers needed.
num1 = int(input("Enter a number"))
#if input's remainder is equal to zero, it is even
# "%" remainder
if num1%2 == 0:
    print("The integer",num1,"is an even number.")
#else function to give an odd number if number did not equal 0.
else:
    print("The integer",num1,"is an odd number.")
```
