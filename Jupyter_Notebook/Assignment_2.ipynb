```python
#program 1
num1= float(input("Enter First Number:"))      
num2= float(input("Enter Second Number:"))
options = int(input("| Enter 1 to Add | Enter 2 to Subtract | Enter 3 to Multiply | Enter 4 to Divide |"))
#if and elif statements set for each operator. 
if options==1:
    answ=num1+num2
elif options==2:
    answ=num1-num2
elif options==3:
    answ=num1*num2
elif options==4:
    answ=num1/num2

print("Your Answer is",answ)
```

```python
#program 4:using membership, y for. 
word = ("cheese")

wordlist = (list(word))
#made the given string into a list in order to compate with the vowels. 
vowelcheck=["a","e","i","o","u"]
total = 0
#checks for vowels in wordlist
#count = wordlist.count, made to cound the number of times "vowelcheck" showed in "word".
#total +- is to add up the number of times it occured overall. 
for x in vowelcheck:
    if x in wordlist:
        count = wordlist.count(x)
        total += count
        print(total)
```

```python
#program 6: using functions
# defined a function to calculate the price total after discount.
def totalPrice(pricetot):
    if pricetot >100:
        discount= round(pricetot* .25, 2)
        realcost= pricetot-discount
        return realcost
         
    elif pricetot >=51 and pricetot <=100:
        discount= round(pricetot* .20, 2)
        realcost= pricetot-discount
        return realcost
        
    elif pricetot >=21 and pricetot <=50:
        discount= round(pricetot* .10, 2)
        realcost= pricetot-discount
        return realcost
        
    elif pricetot <20:
        discount= pricetot*0
        realcost= pricetot-discount
        return realcost
#set code to break from loop if given price is more than 0.         
while True:
    pricetot= float(input("Please enter cost of items purchased: "))
    if pricetot>0:
        break
    else:
        print("Invalid Input, Try Again.")
#used fucntion to determine real price after discount. 
result=totalPrice(pricetot)

print("Total Price After Discount:", result)
```
